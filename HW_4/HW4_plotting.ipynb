Create 9 random plots in a figure (Hint: There is a numpy function for generating random data). 

The top three should be scatter plots (one with green dots, one with purple crosses, and one with blue triangles. Keep in mind the range and conditions for the trignometric functions.

##### All these plots should be on the same figure and not 9 independent figures.

```python
import random
```

```python
f, ax = plt.subplots(nrows=3,ncols=3)

x = np.random.randn(25)
y = np.random.randn(25)

x_trig = np.linspace(-2*np.pi,2*np.pi,50)
y_sin = np.sin(x_trig)
y_cos = np.cos(x_trig)
x_tan = np.linspace(0, np.pi, 12)
y_tan = np.tan(x_tan)

ax[0,0].scatter(x, y, c = 'g')
ax[0,1].scatter(x, y, marker = '+', c = 'purple')
ax[0,2].scatter(x, y, marker = '^')
ax[1,0].plot(np.sort(x),np.sort(y))
ax[1,1].barh(np.sort(x),np.sort(np.abs(y)), ec= 'white')
ax[1,2].hist(x, ec = 'white')
ax[2,0].plot(x_trig, y_sin)
ax[2,1].plot(x_trig, y_cos)
ax[2,2].plot(x_tan, y_tan)
f.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show();
```

#### Q 4.
The data files needed: 

`google_data.txt, ny_temps.txt & yahoo_data.txt`

Use your knowledge with `Python, NumPy, pandas and matplotlib` to reproduce the plot below:

![](answer.png?raw=true)

```python
google = pd.read_csv("google_data.txt", delimiter="\t")
ny = pd.read_csv("ny_temps.txt", delimiter="\t")
yahoo = pd.read_csv("yahoo_data.txt", delimiter="\t")

google_x = google['Modified Julian Date']
google_y = google['Stock Value']
ny_x = ny['Modified Julian Date']
ny_y = ny['Max Temperature']
yahoo_x = yahoo['Modified Julian Date']
yahoo_y = yahoo['Stock Value']


fig, ax1 = plt.subplots()
ax1.plot(google_x, google_y, c = 'green', label = 'Google Stock Value')
ax1.plot(yahoo_x, yahoo_y, c = 'purple', label = 'Yahoo! Stock Value')
ax1.set_xlabel('Date (MJD)')
ax1.set_ylabel('Value(Dollars)', color='purple')
ax1.tick_params('y', colors='green')


t = u"\u00b0" #degree symbol

ax2 = ax1.twinx()
ax2.plot(ny_x, ny_y, c = 'b', linestyle= 'dashed', label= 'NY Mon. High Temp')
ax2.set_ylabel('Temperature(' + t +'F)', color='b')
ax2.tick_params('y', colors='b')
ax2.set_ylim(-150, 100)



lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='center left', fontsize = 'x-small', frameon = False)

plt.title('New York Temperature, Google, and Yahoo!')

plt.show()
```
